#### <i>1.</i> <u><i>Movies Analysis without using Pandas</u>

```python
def calculate_rating_stats(data, industry=None):
    ratings = []
    for row in data:
        if row[3] != 'NULL' and (not industry or row[1] == industry):
            ratings.append(float(row[3]))
    return round(max(ratings),1), round(min(ratings),1), round(sum(ratings)/len(ratings),1)
```

```python
with open("movies.csv") as f:
    data = list(csv.reader(f))
    header = data[0]
    data = data[1:]
    print(header)
```

```python
Max, Min, Avg = calculate_rating_stats(data)
print(color.UNDERLINE + color.BOLD + f"All movies: Maximum rating - {Max}, Minimum rating - {Min}, Average rating - {Avg}")
```

```python
Max, Min, Avg = calculate_rating_stats(data, "Bollywood")
print(color.UNDERLINE + color.BOLD + f"Bolly movies: Maximum rating - {Max}, Minimum rating - {Min}, Average rating - {Avg}")
```

```python
Max, Min, Avg = calculate_rating_stats(data, "Hollywood")
print(color.UNDERLINE + color.BOLD + f"Holly movies: Maximum rating - {Max}, Minimum rating - {Min}, Average rating - {Avg}")
```
